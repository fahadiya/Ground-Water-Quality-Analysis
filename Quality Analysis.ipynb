```python
numeric_object_columns = ['pH', 'CO3', 'SO4', 'NO3', 'PO4', 'F', 'SiO2', 'U(ppb)']
for col in numeric_object_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
converted_info=df.info()
converted_head=df.head()

converted_info
converted_head
```

```python
#mean=60%
missing_value=df.isnull().mean()*100
missing_value
```

```python
columns_to_drop = ['Well ID', 'Year', 'S.No', 'BLOCK', 'u(ppb)', 'LOCATION']
df_cleaned = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
df_cleaned.info()
```

```python
data_cleaned = df_cleaned.fillna(df_cleaned.mean(numeric_only=True))
data_cleaned.isnull().sum()
```

```python
data_cleaned.head()
```

```python
outlier_columns = ['pH', 'EC', 'CO3', 'HCO3', 'Cl', 'SO4', 'NO3', 'PO4', 
                   'TH', 'Ca', 'Mg', 'Na', 'K', 'F', 'SiO2', 'TDS']
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))

plt.suptitle("Box Plot of Water Quality Parameters (Before Outlier Removal)", fontsize=16)
for i, col in enumerate(outlier_columns, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(data=data_cleaned, y=col)
    plt.title(col)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
outlier_counts = {}

for col in outlier_columns:
    Q1 = data_cleaned[col].quantile(0.15)
    Q3 = data_cleaned[col].quantile(0.85)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data_cleaned[(data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

outlier_counts
```

```python
for col in outlier_columns:
    # data_cleaned = data_cleaned[(data_cleaned[col] >= lower_bound) & (data_cleaned[col] <= upper_bound)]
    data_cleaned = data_cleaned[data_cleaned[col].between(lower_bound, upper_bound)]

print("Outlier counts per column:", outlier_counts)

print("Shape after outlier removal:", df_cleaned.shape)
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# List of key water quality parameters
key_params = ['pH', 'EC', 'CO3', 'HCO3', 'Cl', 'SO4', 'NO3', 'PO4', 
                   'TH', 'Ca', 'Mg', 'Na', 'K', 'F', 'SiO2', 'TDS']

# Create histograms for key parameters
plt.figure(figsize=(18, 12))
plt.suptitle("Histogram of Key Water Quality Parameters", fontsize=16)

for i, col in enumerate(key_params, 1):
    plt.subplot(4, 4, int(i))
    sns.histplot(data_cleaned[col], kde=True)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
# Create box plots for key parameters by region (assuming there's a 'Region' or 'State' column)
plt.figure(figsize=(18, 12))
plt.suptitle("Box Plot of Numerical ", fontsize=16)

region='STATE'
for i, col in enumerate(key_params, 1):
    plt.subplot(4, 4, int(i))
    sns.boxplot(data=data_cleaned, y=col)
    plt.title(f"{col} by {region}")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
